```python
import os
import re

def sanitize_text(text):
    # Lowercasing
    text = text.lower()

    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)

    # Remove special characters
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)

    # Remove URLs
    text = re.sub(r'http\S+', '', text)

    # Remove email addresses
    text = re.sub(r'\S+@\S+', '', text)

    # Remove multiple spaces
    text = re.sub(r'\s+', ' ', text)

    return text

def sanitize_directory(input_directory, output_directory):
    # Create output directory if it doesn't exist
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    # Loop through all files in the input directory
    for filename in os.listdir(input_directory):
        if filename.endswith(".txt"):
            input_file_path = os.path.join(input_directory, filename)
            output_file_path = os.path.join(output_directory, filename)

            # Read content from the input file
            with open(input_file_path, 'r', encoding='utf-8') as file:
                text = file.read()

            # Sanitize the text
            sanitized_text = sanitize_text(text)

            # Write sanitized text to the output file
            with open(output_file_path, 'w', encoding='utf-8') as file:
                file.write(sanitized_text)

            print(f"Sanitized '{filename}' and saved to '{output_file_path}'")

if __name__ == "__main__":
    input_directory = r"C:\Users\rafa0\Desktop\pj\laplace\LAPLACE\gutenberg_books"  # Change this to the directory containing your input .txt files
    output_directory = "sanitized_files"  # Directory to save sanitized files

    sanitize_directory(input_directory, output_directory)
```
